```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Dropdown

def plot_eigenfunctions(n, boundary_type):
    """
    Plot eigenfunctions for specified eigenvalue and boundary condition
    """
    
    # Create x-axis data
    x = np.linspace(0, 1, 1000)
    
    # Calculate eigenfunction based on boundary condition
    if boundary_type == 'A':  # Both ends fixed
        eigenfunction = np.sin(n * np.pi * x)
        title = f'Both ends fixed: u(x) = sin({n}Ï€x)'
        
    elif boundary_type == 'B':  # Both ends free
        eigenfunction = np.cos(n * np.pi * x)
        title = f'Both ends free: u(x) = cos({n}Ï€x)'
        
    elif boundary_type == 'C':  # Fixed at left, free at right
        eigenfunction = np.sin((n - 0.5) * np.pi * x)
        title = f'Fixed left, free right: u(x) = sin({n-0.5}Ï€x)'
        
    elif boundary_type == 'D':  # Free at left, fixed at right
        eigenfunction = np.cos((n - 0.5) * np.pi * x)
        title = f'Free left, fixed right: u(x) = cos({n-0.5}Ï€x)'
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot eigenfunction
    ax.plot(x, eigenfunction, 'b-', linewidth=2)
    ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5, alpha=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('u(x)', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_ylim([-1.2, 1.2])
    
    # Add boundary condition markers
    if boundary_type in ['A', 'C']:  # Left end fixed
        ax.plot(0, 0, 'ro', markersize=10, label='Fixed')
    else:  # Left end free
        ax.plot(0, 0, 'go', markersize=10, label='Free')
        
    if boundary_type in ['A', 'D']:  # Right end fixed
        ax.plot(1, 0, 'ro', markersize=10)
    else:  # Right end free
        ax.plot(1, 0, 'go', markersize=10)
    
    # Add legend
    ax.legend()
    
    plt.tight_layout()
    plt.show()

def plot_all_boundaries(n):
    """Plot eigenfunctions for all four boundary conditions simultaneously"""
    
    x = np.linspace(0, 1, 1000)
    
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()
    
    # Calculate eigenfunctions for all four boundary conditions
    eigenfunctions = [
        np.sin(n * np.pi * x),  # A: Both ends fixed
        np.cos(n * np.pi * x),  # B: Both ends free
        np.sin((n - 0.5) * np.pi * x),  # C: Fixed left, free right
        np.cos((n - 0.5) * np.pi * x)   # D: Free left, fixed right
    ]
    
    titles = [
        f'Both ends fixed: u(x) = sin({n}Ï€x)',
        f'Both ends free: u(x) = cos({n}Ï€x)',
        f'Fixed left, free right: u(x) = sin({(n-0.5)}Ï€x)',
        f'Free left, fixed right: u(x) = cos({(n-0.5)}Ï€x)'
    ]
    
    colors = ['b', 'r', 'g', 'm']
    
    for i, (ax, func, title, color) in enumerate(zip(axes, eigenfunctions, titles, colors)):
        ax.plot(x, func, color=color, linewidth=2)
        ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5, alpha=0.5)
        ax.grid(True, alpha=0.3)
        ax.set_title(title, fontsize=11)
        ax.set_xlabel('x')
        ax.set_ylabel('u(x)')
        ax.set_ylim([-1.2, 1.2])
    
    plt.suptitle(f'Eigenfunctions for n = {n} with Different Boundary Conditions', 
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Create interactive interface
print("Eigenfunction Visualization Tool")
print("-" * 40)

# Create widgets
n_slider = IntSlider(
    value=1,
    min=1,
    max=10,
    step=1,
    description='Eigenvalue n:',
    continuous_update=False
)

boundary_dropdown = Dropdown(
    options=[
        ('Both ends fixed', 'A'),
        ('Both ends free', 'B'),
        ('Fixed left, free right', 'C'),
        ('Free left, fixed right', 'D'),
        ('Show all four', 'All')
    ],
    value='A',
    description='Boundary:'
)

# Create interactive plot
@interact(n=n_slider, boundary_type=boundary_dropdown)
def interactive_plot(n, boundary_type):
    if boundary_type == 'All':
        plot_all_boundaries(n)
    else:
        plot_eigenfunctions(n, boundary_type)
```

```python
# æœ€ç»ˆç‰ˆï¼šç®€æ´çš„æ³¢åŠ¨æ–¹ç¨‹å¯è§†åŒ–
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
from IPython.display import HTML, display, clear_output
import matplotlib.animation as animation
import warnings
warnings.filterwarnings('ignore')

# åˆ›å»ºå›¾å½¢ï¼ˆåªåˆ›å»ºä¸€æ¬¡ï¼‰
fig, ax = plt.subplots(figsize=(11, 6.5))
plt.close(fig)  # å…³é—­ä»¥é˜²æ­¢åœ¨å•å…ƒæ ¼ä¸­æ˜¾ç¤ºé™æ€å›¾åƒ

class FinalWaveVisualizer:
    """ç®€æ´çš„æ³¢åŠ¨æ–¹ç¨‹å¯è§†åŒ–"""
    
    def __init__(self):
        # ç‰©ç†å‚æ•°
        self.L = 1.0      # é•¿åº¦
        self.c = 1.0      # æ³¢é€Ÿ
        self.N_max = 20   # æœ€å¤§é¡¹æ•°
        
        # ç©ºé—´ç½‘æ ¼
        self.x = np.linspace(0, self.L, 250)
        
        # åˆå§‹ä½ç§»
        self.f = lambda x: x * (1 - x)
        
        # é¢„è®¡ç®—æ‰€æœ‰æ•°æ®
        self._precompute()
        
        # å›ºå®šyè½´èŒƒå›´ï¼ˆæ ¹æ®æœ€å¤§æŒ¯å¹…ç¡®å®šï¼‰
        self._calculate_fixed_y_range()
        
        # åŠ¨ç”»å¯¹è±¡
        self.anim = None
        
    def _precompute(self):
        """é¢„è®¡ç®—æ‰€æœ‰æ•°æ®ä»¥æé«˜æ€§èƒ½"""
        # æœ¬å¾å€¼
        self.n_vals = np.arange(1, self.N_max + 1)
        self.lambdas = (self.n_vals - 0.5) * np.pi / self.L
        
        # é¢„è®¡ç®—æ‰€æœ‰æœ¬å¾å‡½æ•°
        self.X_mat = np.array([np.sin(lam * self.x) for lam in self.lambdas])
        
        # è®¡ç®—å‚…é‡Œå¶ç³»æ•°
        self.a_n = np.zeros(self.N_max)
        for i in range(self.N_max):
            integrand = self.f(self.x) * self.X_mat[i]
            self.a_n[i] = 2.0 * np.trapezoid(integrand, self.x) / self.L
    
    def _calculate_fixed_y_range(self):
        """è®¡ç®—å›ºå®šçš„yè½´èŒƒå›´"""
        # è®¡ç®—æ‰€æœ‰å¯èƒ½çš„æœ€å¤§æŒ¯å¹…ï¼ˆç”¨20é¡¹å åŠ ï¼‰
        t_samples = np.linspace(0, 4, 100)
        max_amplitude = 0
        
        for t in t_samples:
            u = self.compute_wave_fast(t, self.N_max)
            current_max = np.max(np.abs(u))
            if current_max > max_amplitude:
                max_amplitude = current_max
        
        # è®¾ç½®å›ºå®šçš„yè½´èŒƒå›´
        self.y_min = -max_amplitude * 1.2
        self.y_max = max_amplitude * 1.2
        
        # ç¡®ä¿åŒ…å«åˆå§‹ä½ç§»çš„æœ€å¤§å€¼
        initial_max = np.max(np.abs(self.f(self.x)))
        if initial_max * 1.2 > self.y_max:
            self.y_max = initial_max * 1.2
            self.y_min = -initial_max * 1.2
    
    def compute_wave_fast(self, t, N):
        """å¿«é€Ÿè®¡ç®—æ³¢å½¢ï¼ˆå‘é‡åŒ–ï¼‰"""
        if N <= 0:
            return np.zeros_like(self.x)
        
        N = min(N, self.N_max)
        
        # å‘é‡åŒ–è®¡ç®—
        time_factors = np.cos(self.c * self.lambdas[:N] * t)
        u = np.sum(self.a_n[:N, np.newaxis] * self.X_mat[:N] * time_factors[:, np.newaxis], axis=0)
        
        return u
    
    def create_animation(self, N):
        """åˆ›å»ºåŠ¨ç”»"""
        # æ¸…é™¤ä¹‹å‰çš„å›¾å½¢
        ax.clear()
        
        # è®¾ç½®å›ºå®šçš„åæ ‡è½´èŒƒå›´
        ax.set_xlim(0, self.L)
        ax.set_ylim(self.y_min, self.y_max)
        
        # æ·»åŠ æ°´å¹³å‚è€ƒçº¿ï¼ˆxè½´ï¼‰
        ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3, linewidth=0.8)
        
        # è®¾ç½®åæ ‡è½´æ ‡ç­¾
        ax.set_xlabel('Position x', fontsize=12)
        ax.set_ylabel('Displacement u(x,t)', fontsize=12)
        
        # ç½‘æ ¼
        ax.grid(True, alpha=0.2, linestyle='--')
        
        # ç»˜åˆ¶åˆå§‹ä½ç§»ï¼ˆè™šçº¿ï¼‰
        ax.plot(self.x, self.f(self.x), 'k--', 
                linewidth=1.5, alpha=0.5, label='Initial u(x,0)')
        
        # åˆ›å»ºæ³¢å½¢çº¿æ¡
        wave_line, = ax.plot([], [], 'b-', linewidth=2.5, label=f'Current (N={N})')
        
        # è¾¹ç•Œæ¡ä»¶æ ‡è®°
        # å·¦ç«¯å›ºå®šï¼šçº¢è‰²åœ†ç‚¹æ ‡è®°å›ºå®šç‚¹
        ax.plot(0, 0, 'ro', markersize=10, markerfacecolor='red', 
                markeredgecolor='darkred', label='Fixed end (u=0)')
        
        # å³ç«¯è‡ªç”±ï¼šç»¿è‰²ç®­å¤´æ ‡è®°è‡ªç”±ç«¯
        ax.plot(self.L, 0, 'g^', markersize=10, markerfacecolor='green',
                markeredgecolor='darkgreen', label='Free end (u_x=0)')
        
        # æ·»åŠ å›ºå®šç‚¹è¿çº¿ï¼ˆä»xè½´åˆ°æ³¢å½¢ï¼‰
        fixed_line, = ax.plot([0, 0], [0, 0], 'r-', linewidth=1, alpha=0.5)
        
        # æ·»åŠ å›¾ä¾‹
        ax.legend(loc='upper right', fontsize=9)
        
        # åŠ¨ç”»å¸§æ•°
        frames = 200
        
        # åˆå§‹åŒ–å‡½æ•°
        def init():
            wave_line.set_data([], [])
            fixed_line.set_data([0, 0], [0, 0])
            return wave_line, fixed_line
        
        # åŠ¨ç”»æ›´æ–°å‡½æ•°
        def animate(i):
            # æ—¶é—´èŒƒå›´ï¼š0åˆ°4ç§’
            t = i * 4.0 / frames
            
            # è®¡ç®—æ³¢å½¢
            u = self.compute_wave_fast(t, N)
            
            # æ›´æ–°æ³¢å½¢çº¿æ¡
            wave_line.set_data(self.x, u)
            
            # æ›´æ–°å›ºå®šç‚¹è¿çº¿
            current_u_at_0 = u[0]
            fixed_line.set_data([0, 0], [0, current_u_at_0])
            
            # æ›´æ–°æ ‡é¢˜
            ax.set_title(f'Wave Equation Solution (Fixed-Fixed Boundary)\n'
                        f'c={self.c}, L={self.L}, t={t:.2f}, N={N} terms', 
                        fontsize=12, pad=15)
            
            return wave_line, fixed_line
        
        # åˆ›å»ºåŠ¨ç”»
        self.anim = animation.FuncAnimation(
            fig, animate, init_func=init,
            frames=frames, interval=35,
            blit=True, repeat=True
        )
        
        return self.anim
    
    def display_visualization(self):
        """æ˜¾ç¤ºå¯è§†åŒ–ç•Œé¢"""
        # æ¸…é™¤è¾“å‡º
        clear_output(wait=True)
        
        # æ˜¾ç¤ºè¯´æ˜
        print("=" * 65)
        print("WAVE EQUATION VISUALIZATION")
        print("=" * 65)
        print("Equation: u_tt = cÂ² u_xx,  c=1,  L=1")
        print("Boundary: u(0,t)=0 (fixed left),  u_x(L,t)=0 (free right)")
        print("Initial: u(x,0)=x(1-x),  u_t(x,0)=0")
        print(f"Y-axis range: [{self.y_min:.3f}, {self.y_max:.3f}] (fixed)")
        print("=" * 65)
        print()
        
        # åˆ›å»ºç®€å•çš„äº¤äº’ç•Œé¢ - åªæœ‰Næ»‘å—
        @interact(
            N=IntSlider(
                value=5,
                min=1,
                max=self.N_max,
                step=1,
                description='Number of terms N:',
                continuous_update=False,
                style={'description_width': 'initial'},
                layout={'width': '500px'}
            )
        )
        def show_animation(N):
            """æ˜¾ç¤ºåŠ¨ç”»"""
            # æ¸…é™¤è¾“å‡ºåŒºåŸŸ
            clear_output(wait=True)
            
            # æ˜¾ç¤ºç®€å•åŠ è½½æç¤º
            print(f"ğŸ”„ Generating animation with N={N} terms...")
            
            # åˆ›å»ºåŠ¨ç”»
            anim = self.create_animation(N)
            
            # æ¸…é™¤åŠ è½½æç¤º
            clear_output(wait=True)
            
            # æ˜¾ç¤ºç®€çŸ­ä¿¡æ¯
            print(f"âœ… Animation ready | N = {N} terms")
            print("-" * 40)
            
            # æ˜¾ç¤ºåŠ¨ç”»
            return HTML(anim.to_jshtml())

# ä¸»ç¨‹åº
def main():
    """ä¸»å‡½æ•°"""
    # åˆ›å»ºå¯è§†åŒ–å™¨
    print("Initializing wave equation visualizer...")
    
    visualizer = FinalWaveVisualizer()
    
    print("Visualization ready!")
    
    # æ˜¾ç¤ºå¯è§†åŒ–ç•Œé¢
    visualizer.display_visualization()

# è¿è¡Œä¸»ç¨‹åº
if __name__ == "__main__":
    main()
```
